# 激活函数

## ReLU

```c++
torch::Tensor x = torch::arange(-8.0, 8.0, 0.1, torch::requires_grad(true));
torch::Tensor y = torch::relu(x);


std::vector a(x.data_ptr<float>(), x.data_ptr<float>() + x.numel());
std::vector b(y.data_ptr<float>(), y.data_ptr<float>() + y.numel());

plt::plot(a, b, {{"label", "relu(x)"}});
plt::title("ReLU");
plt::legend();
plt::save("./test1.png");
plt::show();
auto img1 = im::image("./test1.png");
img1
```

```c++
y.backward(torch::ones_like(x), true/*retain_graph*/);

torch::Tensor z = x.grad();

std::vector c(z.data_ptr<float>(), z.data_ptr<float>() + z.numel());
// std::vector d(y.data_ptr<float>(), y.data_ptr<float>() + y.numel());

plt::plot(a, c, {{"label", "grad"}});
plt::title("grad of ReLU");
plt::legend();
plt::save("./test2.png");
plt::show();
auto img2 = im::image("./test2.png");
img2
```

## Sigmoid

```c++
torch::Tensor x = torch::arange(-8.0, 8.0, 0.1, torch::requires_grad(true));
torch::Tensor y = torch::sigmoid(x);


std::vector a(x.data_ptr<float>(), x.data_ptr<float>() + x.numel());
std::vector b(y.data_ptr<float>(), y.data_ptr<float>() + y.numel());

plt::plot(a, b, {{"label", "sigmoid(x)"}});
plt::title("Sigmoid");
plt::legend();
plt::save("./test3.png");
plt::show();
auto img3 = im::image("./test3.png");
img3
```

```c++
y.backward(torch::ones_like(x), true/*retain_graph*/);

torch::Tensor z = x.grad();

std::vector c(z.data_ptr<float>(), z.data_ptr<float>() + z.numel());
// std::vector d(y.data_ptr<float>(), y.data_ptr<float>() + y.numel());

plt::plot(a, c, {{"label", "grad"}});
plt::title("grad of Sigmoid");
plt::legend();
plt::save("./test4.png");
plt::show();
auto img4 = im::image("./test4.png");
img4
```
